```python
from sqlalchemy import create_engine
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Setup Connection to Database 
engine = create_engine('postgresql://postgres:password@localhost:5432/employee_db')
connection = engine.connect()
```

```python
#Make pull from DB
salaries=pd.read_sql("SELECT * FROM salaries", connection)
salaries.head()
```

```python
#Sort by values
salaries['salary'].value_counts()
```

```python
#Get descriptions
salaries['salary'].describe()
```

```python
#Create a histogram to visualize the most common salary ranges for employees
salaries['salary'].plot(kind='hist',bins=10, figsize=(15,10), color='lightcoral').axvline(salaries['salary'].median(), color='r', linestyle='dashed', linewidth=2)
plt.xlabel("Salaries", weight='bold', size=15)
plt.ylabel("Distribution Frequency", weight='bold', size=15)
plt.title("The Most Common Salaries",weight='bold', size=15)
plt.savefig("Images/hist_salary.png")
plt.show()
```

```python
# Make pull from DB for Titles 
titles = pd.read_sql("SELECT * FROM titles", connection)
titles.head()
```

```python
# Make pull for Employees Table
employees = pd.read_sql("SELECT * FROM employees", connection)
employees.head()
```

```python
titles_rename = titles.rename(columns={'title_id': 'emp_title_id'})
titles_rename.head()
```

```python
# Merge tables on employee number
combined_data = pd.merge(salaries, employees, on="emp_no", how="inner")
combined_data.head()
```

```python
# Manager tables on employee titles
combined_data2 = pd.merge(combined_data, titles_rename, on="emp_title_id", how="inner")
combined_data2.head()
```

```python
# Drop unnecessary columns
revised_data = combined_data2.drop(columns = ["birth_date","hire_date","sex","last_name","first_name","emp_title_id"]) 
revised_data.head()
```

```python
# Groupby Title
grouped_data = combined_data2.groupby("title").mean().drop(columns = "emp_no")

grouped_data
```

```python
# Set index
index_df = grouped_data.reset_index()
index_df
```

```python
#Create a bar chart of average salary by title.
x_axis=index_df["title"]
y_axis=index_df["salary"]
plt.bar(x_axis, y_axis, color='maroon')
plt.xticks(x_axis, rotation=45)
plt.xlabel("Job Title")
plt.ylabel("Avg Salary", weight='bold', size=15)
plt.title("Average Salary by Title", weight='bold', size=15)
#To save image
plt.savefig("Images/avg_salary.png")
plt.show()
```
